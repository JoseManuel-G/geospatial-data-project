### Obtengo las cordenadas de Palo Alto

```python
Palo_Alto = {'type': 'Point', 'coordinates': [-122.1430200, 37.4418800]}
```

### Utilizo la API de foursquare para buscar Starbucks en Palo Alto

```python
tok1 = os.getenv("tok1")
tok2 = os.getenv("tok2")
```

```python
enlace = 'https://api.foursquare.com/v2/venues/explore'
```

```python
parametros = { "client_id" : tok1,
          "client_secret" : tok2,
          "v" : "20180323",
          "ll": f"{Palo_Alto.get('coordinates')[1]},{Palo_Alto.get('coordinates')[0]}",
          "query": "Starbucks",
          "limit" : 100
}
```

```python
resp = requests.get (url = enlace, params = parametros)
data = json.loads(resp.text)
```

```python
decoding_data = data.get("response")
```

```python
decoded = decoding_data.get("groups")[0]
```

```python
Starbucks = decoded.get("items")
```

```python
Starbucks
```

```python
from functools import reduce
import operator
```

```python
mapa_nombre = ["venue","name"]
latitud = ["venue", "location","lat"]
longitud = ["venue","location","lng"]
```

```python
def getFromDict(diccionario,mapa):
    return reduce(operator.getitem,mapa,diccionario)
```

```python
lovemosclaro = []
for diccionario in Starbucks:
    paralista = {}
    paralista["type"] = getFromDict(diccionario,mapa_nombre)
    paralista["latitud"] = getFromDict(diccionario,latitud)
    paralista["longitud"] = getFromDict(diccionario,longitud)
    lovemosclaro.append(paralista)
```

```python
df_starbucks = pd.DataFrame(lovemosclaro)
df_starbucks.head()
```

```python
df_starbucks.to_csv('starbucks')
```
